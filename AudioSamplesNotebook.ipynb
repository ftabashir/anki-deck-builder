```python
from TTS.api import TTS
from IPython.display import Audio

model = 'tts_models/de/thorsten/vits'
tts = TTS(model)
text = 'Hallo, mein name ist thorsten.'
audio_array = tts.tts(text)
Audio(audio_array, rate=tts.synthesizer.output_sample_rate)
```
